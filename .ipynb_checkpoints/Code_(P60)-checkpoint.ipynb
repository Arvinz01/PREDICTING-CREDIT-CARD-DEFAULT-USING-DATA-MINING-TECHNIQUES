## Member 1: Dataset Understanding and Data Preprocessing
- Dataset: default of credit card clients.xls
- Output: cleaned CSVs + train/test splits + sklearn preprocessor pickle

```python
# Install dependency for reading .xls (run once)
!pip install xlrd==2.0.1
```

#### Import

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import joblib
```

#### Load dataset

```python
DATA_PATH = "default of credit card clients.xls"  # put file in same folder

# In the UCI dataset, headers start on row 2
df = pd.read_excel(DATA_PATH, header=1, engine="xlrd")

print("=== RAW DATA OVERVIEW ===")
print("Shape:", df.shape)
print("Columns:", list(df.columns))
df.head(3)
```

#### Data Check

```python
print("\n=== DATA TYPES ===")
print(df.dtypes)

print("\n=== MISSING VALUES ===")
missing = df.isna().sum()
print(missing[missing > 0] if missing.any() else "No missing values found.")

print("\n=== DUPLICATE ROWS ===")
print("Duplicates:", df.duplicated().sum())
```

#### Rename & Standardize Columns
- Target column in this file is typically: 'default payment next month'
- We'll rename it to DEFAULT for consistency.

```python
TARGET_COL_RAW = "default payment next month"

if TARGET_COL_RAW not in df.columns:
    raise ValueError("Target column not found!")

df = df.rename(columns={TARGET_COL_RAW: "DEFAULT"})

# Rename PAY_0 → PAY_1 (common convention)
if "PAY_0" in df.columns:
    df = df.rename(columns={"PAY_0": "PAY_1"})

if "ID" not in df.columns:
    raise ValueError("ID column missing!")

print("=== COLUMNS AFTER RENAMING ===")
df.columns
```

#### Fix inconsistent / invalid categorical codes
- SEX: 1=male, 2=female (already clean)
- EDUCATION: (1=grad school,2=university,3=high school,4=others) but dataset has 0,5,6 -> treat as "Others" (4)
- MARRIAGE: (1=married,2=single,3=others) but dataset has 0 -> treat as "Others" (3)

```python
def clean_education(x):
    return 4 if x in [0, 5, 6] else x

def clean_marriage(x):
    return 3 if x == 0 else x

df["EDUCATION"] = df["EDUCATION"].apply(clean_education)
df["MARRIAGE"] = df["MARRIAGE"].apply(clean_marriage)

print("=== CATEGORY COUNTS (POST-CLEANING) ===")
print("\nSEX:\n", df["SEX"].value_counts().sort_index())
print("\nEDUCATION:\n", df["EDUCATION"].value_counts().sort_index())
print("\nMARRIAGE:\n", df["MARRIAGE"].value_counts().sort_index())
```

#### First Exploratory Data Analysis (Member 1 level)

```python
print("=== TARGET DISTRIBUTION ===")
target_counts = df["DEFAULT"].value_counts().sort_index()
print(target_counts)
print("Default rate (%):", round(100 * target_counts[1] / len(df), 2))

plt.figure(figsize=(5,4))
target_counts.plot(kind="bar")
plt.title("Default Distribution")
plt.xlabel("DEFAULT (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
print("=== NUMERICAL SUMMARY ===")
for col in ["LIMIT_BAL", "AGE"]:
    print(f"\n{col}")
    print(df[col].describe())
```

```python
# Quick check: repayment status columns distribution (PAY_1..PAY_6 or PAY_0..)
pay_cols = [c for c in df.columns if c.startswith("PAY_")]
print("=== REPAYMENT STATUS COLUMNS ===")
print(pay_cols)

if pay_cols:
    print(df[pay_cols[0]].value_counts().sort_index().head(15))
```

#### Feature Grouping

```python
# We'll treat SEX/EDUCATION/MARRIAGE as categorical for OneHotEncoding.
# All other non-target, non-ID columns are numeric.
categorical_cols = ["SEX", "EDUCATION", "MARRIAGE"]

# Many teams drop ID (recommended) because it's just a row identifier
drop_cols = ["ID"]
feature_cols = [c for c in df.columns if c not in drop_cols + ["DEFAULT"]]
numeric_cols = [c for c in feature_cols if c not in categorical_cols]

print("Categorical features:", categorical_cols)
print("Numeric feature count:", len(numeric_cols))
```

#### Train-test split (stratified)

```python
X = df[feature_cols]
y = df["DEFAULT"].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Train default rate:", round(y_train.mean(), 4))
print("Test default rate:", round(y_test.mean(), 4))
```

#### Build preprocessing pipeline (for Member 3 model training)
- Numeric -> StandardScaler
- Categorical -> OneHotEncoder

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
    ]
)

# Fit ONLY on training set to prevent leakage
preprocessor.fit(X_train)

print("Preprocessing complete.")
print("Transformed train shape:", preprocessor.transform(X_train).shape)
```

#### Save outputs for the group

```python
OUT_DIR = "member1_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

df.to_csv(f"{OUT_DIR}/data_cleaned.csv", index=False)

X_train.to_csv(f"{OUT_DIR}/X_train.csv", index=False)
X_test.to_csv(f"{OUT_DIR}/X_test.csv", index=False)
y_train.to_csv(f"{OUT_DIR}/y_train.csv", index=False)
y_test.to_csv(f"{OUT_DIR}/y_test.csv", index=False)

joblib.dump(
    {
        "preprocessor": preprocessor,
        "numeric_cols": numeric_cols,
        "categorical_cols": categorical_cols,
        "feature_cols": feature_cols,
        "drop_cols": drop_cols,
        "target_col": "DEFAULT"
    },
    f"{OUT_DIR}/preprocessor.pkl"
)

print("✅ Files saved to:", OUT_DIR)
os.listdir(OUT_DIR)
```

#### Apply SMOTE for Member 3 to do DATA Mining

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Import SMOTE
from imblearn.over_sampling import SMOTE

# Instantiate SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data
# Make sure to pass the preprocessed X_train if you want SMOTE to run on transformed features
# For simplicity, we'll apply it on the raw X_train features first before preprocessing
# However, typically SMOTE is applied AFTER numerical scaling but BEFORE one-hot encoding.
# Given the current notebook structure, X_train is still raw. Let's apply it on raw X_train.
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print(f"Original training set shape: {X_train.shape}, {y_train.shape}")
print(f"Resampled training set shape: {X_train_smote.shape}, {y_train_smote.shape}")

# 1. Create DataFrames for visualization
# Before SMOTE
df_before = pd.DataFrame({'DEFAULT': y_train})
df_before['Label'] = 'Before SMOTE'

# After SMOTE - Ensure this variable matches your SMOTE output variable
df_after = pd.DataFrame({'DEFAULT': y_train_smote})
df_after['Label'] = 'After SMOTE'

# 2. Combine them for easy plotting
combined_df = pd.concat([df_before, df_after])

# 3. Plot using Seaborn
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Label', hue='DEFAULT', data=combined_df, palette='viridis')

plt.title('Effect of SMOTE on Class Distribution (Training Set)', fontsize=15)
plt.xlabel('Dataset State', fontsize=12)
plt.ylabel('Number of Records', fontsize=12)
plt.legend(title='Default Status', labels=['No Default (0)', 'Default (1)'])

# Add value labels on top of bars
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center', xytext = (0, 9), textcoords = 'offset points')

plt.show()
```

## Member 3: Model Development and Training


 **1. Model Implementation (Baseline)**

**Modeling Strategy & Baseline Implementation**

In this phase, we compare three distinct machine learning approaches to determine which model best captures the underlying patterns of credit default behavior.

* Logistic Regression is employed as a baseline linear model due to its simplicity and high interpretability, allowing for clear insights into how individual financial variables influence default risk.

* Random Forest is used as an ensemble-based approach that combines multiple decision trees to model complex, non-linear relationships and interactions between financial features, often resulting in improved predictive performance.

* Neural Network (MLPClassifier) is incorporated to evaluate the effectiveness of a deep learning–based approach. By utilizing multiple hidden layers, the model is capable of learning higher-level feature representations and capturing more intricate patterns within the data that may not be detected by traditional machine learning models.

All models are trained using the SMOTE-balanced training dataset to address class imbalance and ensure that instances of the Default class are adequately represented during learning, preventing bias toward the majority class.

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

```

**2. Hyperparameter Tuning (Grid Search)**

To move beyond baseline performance, we perform Hyperparameter Tuning. We use GridSearchCV to systematically test different combinations of "parameters" (like the number of trees and depth) to find the configuration that yields the highest F1-Score. This metric is crucial because it balances our ability to catch defaulters (Recall) with our precision.

```python
#LogisticRegression
lr = LogisticRegression(max_iter=1000, random_state=42)

#RandomForest
rf = RandomForestClassifier(random_state=42)

#Mlp
mlp = MLPClassifier(max_iter=500, random_state=42)
```

3. **Train Baseline Models**

To trains each baseline model using the SMOTE-resampled training data.

```python
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
mlp.fit(X_train, y_train)
```

**4. Create Evaluation Function**

This function evaluates a model using five strict performance metrics:

* Accuracy

* Precision

* Recall

* F1-score

* ROC-AUC

Using a function avoids repeated code and ensures all models are evaluated fairly.

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)

    return accuracy, precision, recall, f1, roc_auc
```

**5. Evaluate Baseline Models**

To evaluates all baseline models and stores the results in a table for easy comparison.

```python
baseline_results = pd.DataFrame(
    columns=["Accuracy", "Precision", "Recall", "F1-score", "ROC-AUC"],
    index=["Logistic Regression", "Random Forest", "Neural Network"]
)

baseline_results.loc["Logistic Regression"] = evaluate_model(lr, X_test, y_test)
baseline_results.loc["Random Forest"] = evaluate_model(rf, X_test, y_test)
baseline_results.loc["Neural Network"] = evaluate_model(mlp, X_test, y_test)

baseline_results
```

```python
X_train_processed = preprocessor.transform(X_train_smote)
X_test_processed = preprocessor.transform(X_test)

# Fit the models
lr.fit(X_train_processed, y_train_smote)
rf.fit(X_train_processed, y_train_smote)
mlp.fit(X_train_processed, y_train_smote)

baseline_results = pd.DataFrame({
    "Logistic Regression": evaluate_model(lr, X_test_processed, y_test),
    "Random Forest": evaluate_model(rf, X_test_processed, y_test),
    "Neural Network": evaluate_model(mlp, X_test_processed, y_test)
}).T

baseline_results
```

**6.  Grid Search – Logistic Regression**

This cell performs hyperparameter tuning for Logistic Regression using GridSearchCV.
Different values of C are tested to control model regularization.

```python
lr_params = {
    "C": [0.01, 0.1, 1, 10],
    "solver": ["liblinear"]
}

lr_grid = GridSearchCV(
    LogisticRegression(max_iter=1000),
    lr_params,
    scoring="f1",
    cv=5
)

lr_grid.fit(X_train, y_train)
```

**7. Grid Search – Random Forest**

This cell tunes the Random Forest model by testing different numbers of trees and tree depths.

```python
rf_params = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20]
}

rf_grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    rf_params,
    scoring="f1",
    cv=5
)

rf_grid.fit(X_train, y_train)
```

**8. Grid Search – Neural Network (MLP)**

This cell tunes the Neural Network by testing different network sizes and regularization strengths.

```python
mlp_params = {
    "hidden_layer_sizes": [(50,), (100,), (100, 50)],
    "alpha": [0.0001, 0.001],
    "learning_rate_init": [0.001, 0.01]
}

mlp_grid = GridSearchCV(
    MLPClassifier(max_iter=500, random_state=42),
    mlp_params,
    scoring="f1",
    cv=5
)

mlp_grid.fit(X_train, y_train)
```

## Member 4: Model Evaluation and Application Mockup

**Extract Best Models**

This cell retrieves the best-performing model from each Grid Search.

```python
best_lr = lr_grid.best_estimator_
best_rf = rf_grid.best_estimator_
best_mlp = mlp_grid.best_estimator_
```

**Evaluate Tuned Models**

This cell evaluates the tuned models and compares their performance with baseline models.

```python
tuned_results = pd.DataFrame({
    "Logistic Regression (Tuned)": evaluate_model(best_lr, X_test, y_test),
    "Random Forest (Tuned)": evaluate_model(best_rf, X_test, y_test),
    "Neural Network (Tuned)": evaluate_model(best_mlp, X_test, y_test)
}).T

tuned_results
```

```python
final_comparison = pd.concat(
    [baseline_results, tuned_results],
    axis=0
)

final_comparison
```
